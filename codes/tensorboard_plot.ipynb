```python
#import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#load file as dataframe

version = '2.1f'
df = []
for i in range(5):
    df.append(pd.read_csv(f'PPO_log_{version}\Training_{i+1}_1.csv'))

title = f'train_{version}'

#plot dataframe for every element in df
for i in range(5):
    plt.plot(df[i]['Step'], df[i]['Value'])
plt.legend(['run1', 'run2', 'run3', 'run4', 'run5'])
plt.title(f'{title}')
plt.xlabel('Steps')
plt.ylabel('Reward')

#resize plot

plt.rcParams["figure.figsize"] = (13, 6)

#tight layout
plt.tight_layout()

plt.savefig(f'Images/{title}.png')
plt.show()
```
